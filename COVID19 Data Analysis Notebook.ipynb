```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

```python
corona_dataset_csv = pd.read_csv("covid19_Confirmed_dataset.csv")
corona_dataset_csv.head(10)
```

```python
corona_dataset_csv.shape
```

```python
#drop useless columns
#axis =1 for columns
#inplace = True if you dont want to create a new data frame
corona_dataset_csv.drop(["Lat","Long"], axis = 1, inplace = True)
corona_dataset_csv.head(10)
```

```python
#aggregate the rows by country
corona_dataset_agg = corona_dataset_csv.groupby("Country/Region").sum()
corona_dataset_agg
```

```python
corona_dataset_agg.shape
```

```python
#visualise the data related to different countries
#plot a graph that represents cases every day
corona_dataset_agg.loc["China"].plot()
corona_dataset_agg.loc["India"].plot()
corona_dataset_agg.loc["Italy"].plot()
plt.legend()
```

```python
#to find out the rate of change, we need to take the first derivative of a function
corona_dataset_agg.loc["China"].diff().plot()
```

```python
#in the above graph, we see a spike for china on 11 feb. let us look at the max rate of change (or max infection rate) for china
corona_dataset_agg.loc["China"].diff().max()
```

```python
#find max_infection_rate for all countries
countries = list(corona_dataset_agg.index)
max_infection_rate =[]
for c in countries:
    max_infection_rate.append(corona_dataset_agg.loc[c].diff().max())
corona_dataset_agg["max_infection_rate"] = max_infection_rate
corona_dataset_agg
```

```python
#create a new dataframe with only required columns
corona_data = pd.DataFrame(corona_dataset_agg["max_infection_rate"])
corona_data
```

```python
#import happiness report csv
happiness_report_csv = pd.read_csv("worldwide_happiness_report.csv")
happiness_report_csv.head(10)
```

```python
#let us drop unnecessary columns
useless_cols = ["Overall rank","Score","Generosity","Perceptions of corruption"]
happiness_report_csv.drop(useless_cols, axis =1, inplace = True)
happiness_report_csv.head(10)
```

```python
#change indecies of happiness csv to country as corona data has country as index
happiness_report_csv.set_index("Country or region", inplace = True)
happiness_report_csv.head(10)
```

```python
#get the shapes of both datasets to understand which join to perform
corona_data.shape
```

```python
happiness_report_csv.shape
```

```python
#lets use inner join as happiness csv has less countries compared to corona data
data = corona_data.join(happiness_report_csv, how = "inner")
data.head()
```

```python
data.shape
```

```python
#from above, we see that 143 countries are matching between two data sets and have been joined
#now let us see how each of the columns correlate with each other (our interest is to see how max_infection_rate is correlated to other factors)
data.corr()
```

```python
#let us plot graphs for these correlations to understand it better
xaxis = data["GDP per capita"]
yaxis = data["max_infection_rate"]
sns.regplot(x=xaxis,y=np.log(yaxis))
```

```python
xaxis = data["Social support"]
yaxis = data["max_infection_rate"]
sns.regplot(x=xaxis,y=np.log(yaxis))
```

```python
xaxis = data["Healthy life expectancy"]
yaxis = data["max_infection_rate"]
sns.regplot(x=xaxis,y=np.log(yaxis))
```

```python
xaxis = data["Freedom to make life choices"]
yaxis = data["max_infection_rate"]
sns.regplot(x=xaxis,y=np.log(yaxis))
```

```python
'''According to this data and correlations, max_infected_rate is seen to be increasing with increase in GDP per capita,
Social support, life expectancy, and freedom of choices. This is counter-intuitive, but we must note that the reason
for this might be incorrect or no documentation of covid cases in the less developed countries compared to
developed countries '''
```
